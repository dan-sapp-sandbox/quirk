```python
try:
    df = pd.read_spss('../data/raw/global_attitudes.sav')
except Exception as e:
    print(f"Error loading file: {e}")

df['us_interfere'] = df['us_interfere'].str.lower().str.strip()
unwanted_responses = ["donâ€™t know (do not read)", "refused (do not read)"]
filtered_df = df[~df['us_interfere'].isin(unwanted_responses)]
response_counts = filtered_df['us_interfere'].value_counts()

plt.figure(figsize=(12, 3))
response_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.xlabel('Perception')
plt.ylabel('Frequency')
plt.title('Global Perception of US Interference')
plt.xticks(rotation=0)  # Rotate labels for readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
